```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head(10)
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head(20)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
column_data_types = athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
column_data_types = athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)
```

```python
# Check if any values are null.
null_values = combined_sales_df.isnull().sum()
null_values
```

```python
# Check the data type of each column
column_data_types = combined_sales_df.dtypes
column_data_types
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df["invoice_date"] = pd.to_datetime(combined_sales_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
column_data_types = combined_sales_df.dtypes
column_data_types
```

#### Using `pivot_table`
